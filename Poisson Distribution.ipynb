# Poisson Distribution
### the Poisson distribution is a "discrete" probability distribution that expresses the probability of a given "number of events occurring in a fixed interval of time or space" 
### if these events occur with a "known constant mean rate" and "independently" of the time since the last event.

```python
#poisson's formula
def poisson(lam):
 e=  2.71828 
 po = ((lam**size)*(1/(e**lam)))/math.factorial(size)
 return po

#plotGraph
def plotme(xm , ym, val):
    plt.plot(ym, xm, color = 'yellow')
    plt.bar(ym, xm, width = 0.8, color = ['red', 'green'])
    plt.xlabel('Occurances')
    plt.ylabel('Probability')
    plt.title(val)
    return plt.show()
```
